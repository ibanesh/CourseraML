```python
import numpy as np
import scipy.io as sio
import pandas as pd
import matplotlib.pyplot as plt
from IPython import display
from scipy import misc
%matplotlib inline
```

```python
data = sio.loadmat('ex7data2.mat')
X = data['X']
K =3
```

```python
def findClosestCentroids(X,centroids):
    K = centroids.shape[0]
    dist = np.zeros((X.shape[0],K))
    for i in xrange(K):
        diff = X - centroids[i,:]
        dist[:,i] = np.sum(diff**2,1)
    return dist.argmin(1) #np.argmin(dist,1)
```

```python
initial_centroids = np.array([[3,3],[6,2],[8,5]])
idx = findClosestCentroids(X,initial_centroids)
print idx[0:3]
```

```python
def computeCentroids(X,idx,K):
    centroids = np.zeros((K,X.shape[1]))
    for i in xrange(K):
        centroids[i,:]=np.mean(X[np.argwhere(idx==i).ravel(),:],0)
    return centroids

centroids = computeCentroids(X,idx,K)
print centroids
```

```python
def runkMeans(X,initital_centroids,max_iters,plot=False):
    m = X.shape[0]
    K = initial_centroids.shape[0]
    idx = np.zeros(m)
    centroids = initial_centroids
    #skipped code for plotting the progress, see below
    for i in xrange(max_iters):
        idx = findClosestCentroids(X,centroids)
        centroids = computeCentroids(X,idx,K)

    return idx, centroids

max_iters = 10
initial_centroids = np.array([[3,3],[6,2],[8,5]])
idx,centroids = runkMeans(X,initial_centroids,max_iters)
```
